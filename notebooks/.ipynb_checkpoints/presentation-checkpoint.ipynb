#### Lifecycle Test

Collected tweets for 1 day at 15 min, 30 min, 60 min, 2 hrs, 4 hrs, 8hrs, 16 hrs, 24 hrs

Majority retweets/favorites occurred between 1 and 8 hrs after being created.

### Data Collection
---

1. Each tweet is pulled from streaming API
2. Each tweet is updated every 2 hours through rest API for eight hour lifecycle

Six sets of (150 )tweets collected per day for one week.

### Data Processing
---

1. Influence Score
    * Weighted score composed used to compare tweets.
    * Similar to using retweet_count but adds estimate for user
2. Influence Interval
    * Two-hour block of time relative to created time.

In our system each tweet is represented 5 times. To find any single tweet, you need id_str, day, set, call numbers.

To identify the influence interval, the influence score for each interval is compared and only the interval with the highest score is kept.
